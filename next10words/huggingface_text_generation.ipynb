```python
from transformers import pipeline # Import der Funktion pipeline des Moduls transformers
```

```python
pipe = pipeline("text-generation", model="Qwen/Qwen3-0.6B",max_new_tokens=1, max_length=1, num_return_sequences=5)
```

```python
prompt = "Die Bayerische Staatskanzlei ist eine"
antwortprompt = pipe(prompt)
antwortprompt
```
